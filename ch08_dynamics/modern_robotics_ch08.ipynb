```python
# 라그랑지안 동역학 해석을 위한 2-link manipulator의 param

import numpy as np

# 2-link manipulator 파라미터
L1 = 1.0  # link 1 길이 [m]
L2 = 1.0  # link 2 길이 [m]
m1 = 1.0  # link 1 질량 [kg]
m2 = 1.0  # link 2 질량 [kg]
g = 9.81  # 중력 가속도 [m/s^2]
```

```python
# 각 링크들은 점 질량을 가지며 링크의 끝 지점에 존재한다라고 가정

import numpy as np

def calculate_pos(L1, L2, theta1, theta2):
    
    x1 = L1 * np.cos(theta1)
    y1 = L1 * np.sin(theta1)

    x2 = x1 + L2 * np.cos(theta1 + theta2)
    y2 = y1 + L2 * np.sin(theta1 + theta2)

    pos1 = np.array([x1, y1])
    pos2 = np.array([x2, y2])

    return pos1, pos2
```

```python
def calculate_vel(L1, L2, theta1, theta2, dtheta1, dtheta2):

    vx1 = -L1 * np.sin(theta1) * dtheta1
    vy1 =  L1 * np.cos(theta1) * dtheta1

    vx2 = vx1 - L2 * np.sin(theta1 + theta2) * (dtheta1 + dtheta2)
    vy2 = vy1 + L2 * np.cos(theta1 + theta2) * (dtheta1 + dtheta2)

    vel1 = np.array([vx1, vy1])
    vel2 = np.array([vx2, vy2])

    return vel1, vel2
```

```python
def calculate_lagrangian(L1, L2, m1, m2, g, theta1, theta2, dtheta1, dtheta2):

    pos1, pos2 = calculate_pos(L1, L2, theta1, theta2)
    vel1, vel2 = calculate_vel(L1, L2, theta1, theta2, dtheta1, dtheta2)

    # Kinetic energy
    K1 = 0.5 * m1 * np.dot(vel1, vel1)
    K2 = 0.5 * m2 * np.dot(vel2, vel2)
    K = K1 + K2

    # Potential energy (높이 = y좌표)
    P1 = m1 * g * pos1[1]
    P2 = m2 * g * pos2[1]
    P = P1 + P2

    L = K - P
    return L
```

```python
# 테스트
theta1, theta2 = np.pi/4, np.pi/6
dtheta1, dtheta2 = 1.0, 0.5

L = calculate_lagrangian(L1, L2, m1, m2, g, theta1, theta2, dtheta1, dtheta2)
print(L)
```
